```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

data = pd.read_csv('/content/drive/My Drive/ДИПЛОМ/Colab Notebooks/data_and_klast_new.csv', encoding = 'utf-8')

data = pd.DataFrame(data)
```

```python
# Разделение на 2019 и 2020

mask = data['Year'] == 2019
data_2019, data_2020 = data[mask], data[~mask]
```

```python
# Разделение 2019 на 4 и 5 классы

mask = data_2019['Класс'] == 4
data_2019_4, data_2019_5 = data_2019[mask], data_2019[~mask]
```

```python
# Разделение 2020 на 5 и 6 классы

mask = data_2020['Класс'] == 5
data_2020_5, data_2020_6 = data_2020[mask], data_2020[~mask]
```

```python
# Разделение 2019 / 4 на РУ и МА

mask = data_2019_4['Код_предмета'] == 1
data_2019_4_PY, data_2019_4_MA = data_2019_4[mask], data_2019_4[~mask]
```

```python
# Разделение 2019 / 5 на РУ и МА

mask = data_2019_5['Код_предмета'] == 1
data_2019_5_PY, data_2019_5_MA = data_2019_5[mask], data_2019_5[~mask]
```

```python
# Разделение 2020 / 5 на РУ и МА

mask = data_2020_5['Код_предмета'] == 1
data_2020_5_PY, data_2020_5_MA = data_2020_5[mask], data_2020_5[~mask]
```

```python
# Разделение 2020 / 6 на РУ и МА

mask = data_2020_6['Код_предмета'] == 1
data_2020_6_PY, data_2020_6_MA = data_2020_6[mask], data_2020_6[~mask]
```
